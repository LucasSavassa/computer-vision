### In place

Adding a scalar to a tensor using **add()**:

```python
print(tensor.add(1))
print(tensor)
```

Adding a scalar to a tensor using **add_()**:

```python
print(tensor.add_(1))
print(tensor)
```

Multiplying a scalar to a tensor using **mul()**:

```python
print(tensor.mul(2))
print(tensor)
```

Multiplying a scalar to a tensor using **mul_()**:

```python
print(tensor.mul_(2))
print(tensor)
```

Dividing a scalar to a tensor using **div()**:

```python
print(tensor.div(2))
print(tensor)
```

Dividing a scalar to a tensor using **div_()**:

```python
print(tensor.div_(2))
print(tensor)
```

### Transpose

```python
tensor1 = torch.full((2,3), 2)
tensor2 = torch.full((2,3), 3)
print(tensor1)
print(tensor2)
```

```python
print(tensor1.T)
print(tensor2.T)
```

```python
print(torch.matmul(tensor1, tensor2.T))
print(torch.matmul(tensor1, tensor2))
```

```python
print(torch.matmul(tensor1, tensor2.T))
print(torch.mm(tensor1, tensor2.T))
```

### Linear

```python
torch.manual_seed(1)
```

```python
linear = torch.nn.Linear(2, 1)

input = torch.tensor([2.,2.])
output = linear(input)

print(output)
print(linear.weight)
print(linear.bias)
```

```python
linear = torch.nn.Linear(2, 1)

input = torch.tensor([[2.,2.],[1.,1.]])
output = linear(input)

print(output)
print(linear.weight)
print(linear.bias)
```

### Conversion

```python
tensor = torch.tensor([1,2,3], dtype=torch.float32)
print(tensor)
print(tensor.dtype)
```

```python
tensor2 = tensor.type(torch.int8)
print(tensor2)
print(tensor2.dtype)
```

```python
tensor3 = tensor.type(torch.float16)
print(tensor3)
print(tensor3.dtype)
```

### Aggregation

```python
tensor = torch.arange(0,100,10)
print(tensor)
```

```python
print(f"max : {tensor.max()}")
print(f"index max : {tensor.argmax()}")
print(f"min : {tensor.min()}")
print(f"index min : {tensor.argmin()}")
print(f"sum : {tensor.sum()}")
print(f"mean : {tensor.type(torch.float32).mean()}")
print(f"std : {tensor.type(torch.float32).std()}")
```
