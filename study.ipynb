# Tensor

```python
import torch
```

### Scalar

```python
scalar = torch.tensor(5)
```

```python
print("scalar : " + str(scalar))
print("shape : " + str(scalar.shape))
print("ndim : " + str(scalar.ndim))
```

### Vector

```python
vector = torch.tensor([5,5])
```

```python
print("vector : " + str(vector))
print("shape : " + str(vector.shape))
print("ndim : " + str(vector.ndim))
```

### Matrix

```python
matrix = torch.tensor([[5,5,5],[5,5,5]])
```

```python
print("matrix : " + str(matrix))
print("shape : " + str(matrix.shape))
print("ndim : " + str(matrix.ndim))
```

### Tensor

```python
tensor = torch.tensor([[[255,0,0],[0,255,0],[0,0,255]],[[255,255,0],[255,0,255],[0,255,255]],[[255,255,255],[0,0,0],[127,127,127]]])
```

```python
print("tensor : " + str(tensor))
print("shape : " + str(tensor.shape))
print("ndim : " + str(tensor.ndim))
```

### Indexing

1. Print the elements in the first layer:

```python
elements = tensor[:,:,0]
```

```python
print(elements)
```

2. Print the elements in the first layer and second row:

```python
elements = tensor[1,:,0]
```

```python
print(elements)
```

3. Print the element at the first layer, second row, and third column:

```python
elements = tensor[1,2,0]
```

```python
print(elements)
```

4. Check the numpy representation of a tensor:

```python
print(tensor.numpy())
```

### Plotting

```python
import matplotlib.pyplot as plt
```
